```python
x = df.iloc[:,1:2].values
y = df.iloc[:,2].values
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y,marker='o',linestyle='-', color="red")
plt.xlabel('Level')
plt.ylabel('Salary')
plt.grid(False)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
poly=PolynomialFeatures(degree=3)
```

```python
x_poly=poly.fit_transform(x)
```

```python
x
```

```python
x_poly
```

```python
linear_model=LinearRegression()
```

```python
linear_model.fit(x_poly,y)
```

```python
y1=linear_model.predict(x_poly)
```

```python
plt.scatter(x,y,marker='o',c="red")
plt.plot(x,y1,'green')
plt.xlabel('level')
plt.ylabel('salary')
plt.show()
```
