```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
df = pd.read_csv(r"C:\Users\SURENDRA\Downloads\salary_data.csv")
```

```python
df 
```

```python
x = df.iloc[:,0].values
y = df.iloc[:,1].values
```

```python
type(x)
```

```python
type(y)
```

```python
plt.scatter(x,y,c="#FFDE42")
plt.xlabel('Years of experience')
plt.ylabel('salary')
plt.title('Y.O.E vs Salary')
plt.show()
```

```python
x_bar = np.mean(x)
y_bar = np.mean(y)
x_bar,y_bar
```

```python
numerator = np.sum((x-x_bar)*(y-y_bar))
denominator = np.sum((x-x_bar)**2)
```

```python
teta1 = numerator/denominator
```

```python
teta0 = y_bar-(teta1 * x_bar)
```

```python
teta0
```

```python
y_pred = teta1*x + teta0
```

```python
plt.scatter(x,y,c="#FFDE42")
plt.scatter(x, y_pred, c="#BF4646")
plt.plot(x,y_pred,c="#254F22")
for i in range(len(x)):
    plt.plot([x[i], x[i]],[y[i],y_pred[i]],color="brown")
plt.xlabel('Years of experience')
plt.ylabel('salary')
plt.title('Y.O.E vs Salary')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse = mean_squared_error(y,y_pred)
print(f"The mean squared error is {mse}")
```

```python
r2 = r2_score(y,y_pred)
print(f"The r2 score is {r2}")
```
