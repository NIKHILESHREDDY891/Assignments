```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv("salary_data.csv")
```

```python
X = data["YearsExperience"].values
y = data["Salary"].values

X = np.array(X)
y = np.array(y)
```

```python
plt.scatter(X, y, color="blue", label="Data points")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary vs Experience")
plt.legend()
plt.show()
```

```python
x_mean = np.mean(X)
y_mean = np.mean(y)

# θ1 = Σ(xi - x̄)(yi - ȳ) / Σ(xi - x̄)^2
theta1 = np.sum((X - x_mean) * (y - y_mean)) / np.sum((X - x_mean)**2)

# θ0 = ȳ - θ1 * x̄
theta0 = y_mean - theta1 * x_mean

print(f"θ0 (Intercept): {theta0}")
print(f"θ1 (Slope): {theta1}")
```

```python
y_pred = theta1 * X + theta0
```

```python
plt.figure(figsize=(6,4))
```

```python
plt.scatter(X, y, color="blue", label="Actual data") 
plt.plot(X, y_pred, color="black", marker="o",mfc="green",ls="solid",label="Regression line")
for i in range(len(X)):
    plt.plot([X[i], X[i]], [y[i], y_pred[i]], color='red')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression Fit")
plt.legend()
plt.show()
```

```python
# 8. Measure performance
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R² Score: {r2}")
```
