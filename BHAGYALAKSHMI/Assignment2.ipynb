```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('C:/Users/DT019/Downloads/salary_data.csv')
```

```python
df.head()
```

```python
X = df['YearsExperience'].values
y = df['Salary'].values
```

```python
plt.scatter(X,y,color='black')
plt.title('Salary vs Experience')
plt.xlabel('YearsExperience')
plt.ylabel('salary')
plt.show()
```

```python
x_bar = np.mean(X)
y_bar = np.mean(y)

theta1 = np.sum((X - x_bar)*(y - y_bar)) / np.sum((X - x_bar)**2)
theta0 = y_bar - theta1*x_bar

print("Theta1 (Slope):", theta1)
print("Theta0 (Intercept):", theta0)
```

```python
y_pred = theta1 * X + theta0
```

```python
print(y_pred[:5])
```

```python
plt.scatter(X, y, color='black')
plt.plot(X, y_pred,marker='o',mfc='green')
for i in range(len(X)):
    plt.plot([X[i],X[i]],[y[i],y_pred[i]],color='blue')
plt.title("Simple Linear Regression")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")

plt.show()
```

```python
score=r2_score(y,y_pred)
print("R2 score:",score)
```
