```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:/Users/haris/Downloads/salary_data.csv')
```

```python
X = df['YearsExperience'].values
y = df['Salary'].values
n=len(X)
a = 0
b = 0
alpha=0.01
iterations=10000
cost_history = []
```

```python
for i in range(iterations):
    y_pred = b * X + a
    
    c1 = (1/n) * np.sum((y - y_pred)**2)
    cost_history.append(c1)
    
    d_a = (-2/n) * sum(y - y_pred)
    d_b = (-2/n) * sum(X*(y - y_pred))
    
    a = a - alpha * d_a
    b = b - alpha * d_b
print("a:", a)
print("b:", b)
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, y_pred, color='blue')
plt.title("Linear Regression using Gradient Descent")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, y_pred,marker='o',mfc='green',color='yellow')
plt.vlines(X, y, y_pred, color='black')
plt.title("Linear Regression using Gradient Descent")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(range(iterations),cost_history)
plt.title("Cost Function vs Iterations")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
```
