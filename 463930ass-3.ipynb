```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:/Users/haris/Downloads/salary_data.csv')
```

```python
X = df['YearsExperience'].values
y = df['Salary'].values
n=len(X)
theta0 = 0
theta1 = 0
alpha=0.01
cost = []
```

```python
for i in range(1000):
    y_pred = theta1 * X + theta0
    d_theta0 = (-2/n) * sum(y - y_pred)
    d_theta1 = (-2/n) * sum(X*(y - y_pred))
    theta0 = theta0 - alpha * d_theta0
    theta1 = theta1 - alpha * d_theta1
print("Theta0:", theta0)
print("Theta1:", theta1)
```
