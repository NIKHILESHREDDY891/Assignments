```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv('salary_data.csv')
print(data.head())
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
X = df['YearsExperience'].values
y = df['Salary'].values
```

```python
plt.scatter(X, y)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
plt.show()
```

```python
x_mean = np.mean(X)
y_mean = np.mean(y)

numerator = np.sum((X - x_mean) * (y - y_mean))
denominator = np.sum((X - x_mean) ** 2)

theta1 = numerator / denominator
theta0 = y_mean - theta1 * x_mean

print("Slope (θ1):", theta1)
print("Intercept (θ0):", theta0)
```

```python
y_pred = theta1 * X + theta0
```

```python
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')

plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Simple Linear Regression using Least Squares")
plt.legend()

plt.show()
```

```python
r2 = r2_score(y, y_pred)
print("R2 Score:", r2)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
data = pd.read_csv("salary_data.csv")

# Step 2: Extract X and y
X = data["YearsExperience"].values
y = data["Salary"].values

# Step 3: Calculate means
x_mean = np.mean(X)
y_mean = np.mean(y)

# Step 4: Calculate slope (theta1)
numerator = np.sum((X - x_mean) * (y - y_mean))
denominator = np.sum((X - x_mean) ** 2)
theta1 = numerator / denominator

# Step 5: Calculate intercept (theta0)
theta0 = y_mean - theta1 * x_mean

print("Slope (θ1):", theta1)
print("Intercept (θ0):", theta0)

# Step 6: Predict values
y_pred = theta1 * X + theta0

# Step 7: Plot actual data points
plt.scatter(X, y, color='blue')

# Step 8: Plot regression line
plt.plot(X, y_pred, color='green')

# Step 9: Draw residual (error) lines
for i in range(len(X)):
    plt.plot([X[i], X[i]], [y[i], y_pred[i]], color='red')

# Step 10: Labels and title
plt.title("Salary VS Experience")
plt.xlabel("Year of Experience")
plt.ylabel("Salary")

# Step 11: Show graph
plt.show()
```
