```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
dataset= pd.read_csv(r'C:\Users\DT043\Downloads\salary_data.csv')
print(dataset.head())
```

```python
dataset
```

```python
X = dataset.iloc[:, 0].values
y = dataset.iloc[:, 1].values
```

```python
X
```

```python
plt.scatter(X, y)
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Salary vs Experience")
plt.show()
```

```python
x_mean = np.mean(X)
y_mean = np.mean(y)
```

```python
num = np.sum((X - x_mean) * (y - y_mean))
den = np.sum((X - x_mean)**2)

theta1 = num / den
```

```python
theta0 = y_mean - theta1 * x_mean
```

```python
y_pred = theta1 * X + theta0
```

```python
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='black',marker="s",mfc="GREEN",mec="yellow",ls="solid",lw=2)
for i in range(len(X)):
    plt.plot([X[i],X[i]],[y[i],y_pred[i]],color='red')
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Linear Regression using Least Squares")
plt.show()
```

```python
mse = mean_squared_error(y, y_pred)
print("Mean Squared Error:", mse)
```
