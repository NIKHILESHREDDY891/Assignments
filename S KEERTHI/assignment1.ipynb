```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
dataset= pd.read_csv(r"C:\Users\DT048\Desktop\position_salaries.csv")
type(dataset)
```

```python
dataset
```

```python
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
plt.scatter(X, y,color='red')
plt.title('Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
                            
```

```python
# Predict on test set
y_pred = lin_reg.predict(poly_reg.transform(X))

# Compare predictions
print("Predicted:", y_pred)
print("Actual:",y)
```

```python

plt.scatter(X, y, color='red')
plt.plot(X, lin_reg.predict(poly_reg.transform(X)), color='blue')
plt.title('Poynomial Regression Model')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
