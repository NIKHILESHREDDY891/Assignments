```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
dataset= pd.read_csv(r"C:\Users\DT048\Desktop\salary_data.csv")
print(dataset.head())
```

```python
X = dataset.iloc[: , 0].values
y = dataset.iloc[:, 1].values
```

```python
plt.scatter(X,y)
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("salary vs Experience")
plt.show()
```

```python

x_mean = np.mean(X)
y_mean = np.mean(y)

num = np.sum((X - x_mean) * (y - y_mean))
den = np.sum((X - x_mean)**2)

b1 = num / den
b0 = y_mean - b1 * x_mean

print("Intercept (b0):", b0)
print("Slope (b1):", b1)
```

```python
def predict(x):
    return b0 + b1 * x
y_pred = predict(X)
```

```python
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='black', marker="d", mfc="red", mec="yellow", ls="solid")

plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Linear Regression using Least Squares")

plt.show()
```

```python
mse = mean_squared_error(y, y_pred)
r2 = mean_squared_error(y, y_pred)

print("Mean Squared Error:", mse)
print("R² Score:", r2)
```
